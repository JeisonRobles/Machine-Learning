```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
H = nx.Graph()

nodes = ['Healthy_Subject_1','Healthy_Subject_2','Healthy_Subject_3',
        'Healthy_Subject_4','Healthy_Subject_5','Healthy_Subject_6',
        'Healthy_Subject_7','Healthy_Subject_8','Healthy_Subject_9',
        'Healthy_Subject_10','Healthy_Subject_11','Healthy_Subject_12']

H.add_node('Infected_Subjet_A')
H.add_node('Infected_Subjet_B')
H.add_node('Infected_Subjet_C')
H.add_node('Infected_Subjet_D')
H.add_node('Infected_Subjet_E')

H.add_nodes_from(nodes)


Healthy_Athlete_List = ['Healthy_Athlete_1','Healthy_Athlete_2',
                       'Healthy_Athlete_3','Healthy_Athlete_4',
                       'Healthy_Athlete_5','Healthy_Athlete_6',
                        'Healthy_Athlete_7','Healthy_Athlete_8',
                        'Healthy_Athlete_9','Healthy_Athlete_10',
                       'Healthy_Athlete_11','Healthy_Athlete_12']

H.add_nodes_from(Healthy_Athlete_List)


```

```python
#Adding Healthy conections
'''Of course this can be assigned based on some logic 
in previous stage'''

H.add_edge('Healthy_Subject_1','Healthy_Subject_10', weight=1.7)
H.add_edge('Healthy_Subject_3','Healthy_Subject_11', weight=14.7)
H.add_edge('Healthy_Subject_4','Healthy_Subject_7', weight=14.7)
H.add_edge('Healthy_Subject_5','Healthy_Subject_8', weight=14.7)
H.add_edge('Healthy_Subject_8','Healthy_Subject_2', weight=14.7)
H.add_edge('Healthy_Subject_11','Healthy_Subject_3', weight=14.7)
H.add_edge('Healthy_Subject_12','Healthy_Subject_4', weight=14.7)
H.add_edge('Healthy_Subject_5','Healthy_Subject_9', weight=14.7)
H.add_edge('Healthy_Subject_6','Healthy_Subject_6', weight=14.7)
H.add_edge('Healthy_Subject_8','Healthy_Subject_12', weight=14.7)
H.add_edge('Healthy_Subject_10','Healthy_Subject_12', weight=14.7)
H.add_edge('Healthy_Subject_11','Healthy_Subject_6', weight=14.7)
H.add_edge('Healthy_Subject_3','Healthy_Subject_4', weight=14.7)
H.add_edge('Healthy_Subject_3','Healthy_Subject_4', weight=14.7)

#Adding Infected edges
H.add_edge('Infected_Subjet_A','Infected_Subjet_B', weight=14.7)
H.add_edge('Infected_Subjet_A','Infected_Subjet_C', weight=14.7)
H.add_edge('Infected_Subjet_A','Infected_Subjet_D', weight=14.7)
H.add_edge('Infected_Subjet_A','Infected_Subjet_E', weight=14.7)


#Joining Infected and health
H.add_edge('Healthy_Subject_3','Infected_Subjet_E', weight=14.7)


#Adding Athletes
H.add_edge('Healthy_Athlete_1','Healthy_Athlete_2', weight = 10)
H.add_edge('Healthy_Athlete_1','Healthy_Athlete_3', weight = 10)
H.add_edge('Healthy_Athlete_1','Healthy_Athlete_4', weight = 10)
H.add_edge('Healthy_Athlete_4','Healthy_Athlete_2', weight = 10)
H.add_edge('Healthy_Athlete_2','Healthy_Athlete_5', weight = 10)
H.add_edge('Healthy_Athlete_5','Healthy_Athlete_6', weight = 10)
H.add_edge('Healthy_Athlete_6','Healthy_Athlete_3', weight = 10)
H.add_edge('Healthy_Athlete_6','Healthy_Athlete_7', weight = 10)
H.add_edge('Healthy_Athlete_7','Healthy_Athlete_8', weight = 10)
H.add_edge('Healthy_Athlete_8','Healthy_Athlete_9', weight = 10)
H.add_edge('Healthy_Athlete_7','Healthy_Athlete_9', weight = 10)
H.add_edge('Healthy_Athlete_9','Healthy_Athlete_10', weight = 10)
H.add_edge('Healthy_Athlete_10','Healthy_Athlete_11', weight = 10)
H.add_edge('Healthy_Athlete_10','Healthy_Athlete_12', weight = 10)
H.add_edge('Healthy_Athlete_12','Healthy_Athlete_1', weight = 10)

#Joinning Healthy-Athlete

H.add_edge('Healthy_Athlete_8','Healthy_Subject_9', weight = 1)
```

```python
#Get nodes list
H.nodes
```

```python
#Gets edge list
H.edges
```

```python
#Assign colors depending on the value of node
color_map = []
for node in H:
    texto = str(node)
    if texto.find("Infected") == -1:
        if texto.find("Healthy_Athlete") ==-1:
            color_map.append('teal')
        else:
            color_map.append('green')
        

    else: 
        
        color_map.append('red') 
```

```python
#Fills a list of sizes based on an specific value
node_sizea = []

for node in H:
    texto = str(node)
    if texto.find("Infected_Subjet_A") != -1:
        node_sizea.append(700)
    else:
        node_sizea.append(700)
```

```python
#Script for plotting the graph
nx.draw(H, with_labels=True, 
        node_color = color_map, 
        edge_color = 'blue',

       node_size = node_sizea)
plt.title("Simple Infection Network", fontsize =100)
plt.rcParams["figure.figsize"] = (15,15)
```
